### Q1 Number of multiplications and additions required for a matrix multiplication of [A]n X k and [B]k X m

Number of additions required are n* m* (k-1) and number of multiplications required are n* m* k.
The number of elements in the resultant matrix are n* m, therefore the operations are a multiple of n* m.
For every element in the resultant matrix, the operations are done between k pairs of numbers from the two matrices, therefore k pairs are multiplied together. These k pairs are then added together requiring k-1 addition operations.

### Q2 Python code and numpy to multiply matrices

```python
#Python code for matrix multiplication
def matrix_multiply(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("not possible")
    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    return result
```

```python
#Using numpy for matrix multiplication
import numpy as np

def matrix_multiply_numpy(matrix1, matrix2):
    try:
        result = np.dot(matrix1, matrix2)
        return result
    except ValueError:
        print("not possible")
```

```python
matrix_a =([[1, 2, 3],
           [4, 5, 6],
           [4, 5, 6],
           [4, 5, 6],
           [4, 5, 6],
           [4, 5, 6]])

matrix_b = ([[7, 8,10,11,12,13,14],
            [9, 10,5,6,7,8,1],
            [11, 12,4,7,2,6,2]])

%time matrix_multiply_numpy(matrix_a, matrix_b)
```

```python
%time matrix_multiply(matrix_a, matrix_b)
```

### Q3 Time complexity of finding median

Time complexity required for finding median using largest element each time requires a O(n*n/2) which is O(n^2). This is because n/2 passes are required for finding the median from n elements array. Also a space complexity of O(n) is required to store the largest elements found till the median is found.

##### Better method

```python
# Passing method
def find_median_passing(arr):
    n = len(arr)
    largest_elements = set()
    
    for _ in range(n // 2):
        maxi = float('-inf')
        for num in arr:
            if num > maxi and num not in largest_elements:
                maxi = num
        
        if maxi != float('-inf'):
            largest_elements.add(maxi)

    if n % 2 == 1:
        return min(largest_elements)
    else:
        return (min(largest_elements) + max(largest_elements)) / 2
```

A better approch can be sorting the array requiring a time complexity of O(n logn) and then picking up the middle element or the avarage of the middle elements if two. If a copy of the array is created a space complexity of O(n) is required.

```python
# Sort and find median
def find_median_sort(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)
    if n % 2 == 1:
        return sorted_arr[n // 2]
    else:
        mid1 = n // 2 - 1
        mid2 = n // 2
        return (sorted_arr[mid1] + sorted_arr[mid2]) / 2
        
```

### Q4 Gradient of funtion (x^2)y + (y^3)sin(x)

Differentiating function with respect to x gives:  (2*x*y) + (y^3)* cos(x). 
Differentiating with respect to y gives: x^2 + 3* (y^2)*sin(x)

### Q5 Using JAX to confirm the gradient

```python
import jax
import jax.numpy as jnp
import numpy as np

# Define the function using JAX-compatible numpy
def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

# Define the gradient function using JAX
grad_fn = jax.grad(f, (0, 1))

# Generate random values for x and y
random_values = np.random.rand(5, 2)  # 5 samples, 2 variables

# Compute gradients using JAX
jax_gradients = np.array([grad_fn(x, y) for x, y in random_values])

# Display the results
for i in range(len(random_values)):
    x_val, y_val = random_values[i]
    grad_x, grad_y = jax_gradients[i]
    print(f"values: ({x_val}, {y_val}), Gradient: {grad_x}, {grad_y}")
```

### Q6 Using Sympy to calculate the Derivatives and gradient

```python
import sympy as sp
x, y = sp.symbols('x y')

f = x**2 * y + y**3 * sp.sin(x)
derivative_x = sp.diff(f, x)
derivative_y = sp.diff(f, y)
```

### Q10 Matrix of size 20 X 5

```python
import pandas as pd
matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

# Find the column with the highest standard deviation
max_std_col = df.std().idxmax()
# Find the row with the lowest mean
min_mean_row = df.mean(axis=1).idxmin()

print("DataFrame:")
print(df)

# Print the result
print("\nColumn with the highest standard deviation:", max_std_col)
print("Row with the lowest mean:", min_mean_row)
```

### Q11 Column 'f'

```python
# Add a new column "f" 
df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

# Add a new column "g" 
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

# Find the number of rows where the value in the column "g" is "LT8"
num_lt8_rows = len(df[df['g'] == 'LT8'])

# Find the standard deviation of column "f" for the rows where "g" is "LT8" and "GT8"
std_dev_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_dev_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

print("Modified DataFrame:")
print(df)
print("\nNumber of rows where the value in column 'g' is 'LT8':", num_lt8_rows)
print("Standard deviation of column 'f' for rows where 'g' is 'LT8':", std_dev_lt8)
print("Standard deviation of column 'f' for rows where 'g' is 'GT8':", std_dev_gt8)
```

### Q13 Argmin of a numpy array

```python
def custom_argmin(arr):
    if len(arr) == 0:
        raise ValueError("Input array is empty")
    min_value = arr[0]
    min_index = 0

    for i, value in enumerate(arr):
        if value < min_value:
            min_value = value
            min_index = i

    return min_index, min_value
```

```python
arr = np.array([4, 1, 8, 3, 6,3524,-123])
custom_argmin_result, min_value = custom_argmin(arr)

np_argmin_result = np.argmin(arr)

print("Custom argmin result (index, value):", custom_argmin_result, min_value)
print("np.argmin result:", np_argmin_result)
```
